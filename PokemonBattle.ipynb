```python
"""
Group 7: Ariana Lalehparvar, Fei Zheng,

Implement Class:

Develop a Pokémon class with attributes like name, type, health, defense, and moves.
Encapsulate Behavior:

Implement member functions within the class to represent actions like performing an attack.
Manage Object Interaction:

Facilitate interactions between Pokémon objects to allow for a battle between two to occur.
Simulate Game Logic:

Implement a simple battle simulation where Pokémon objects take turns to perform moves until one runs out of health.
"""
import random
class Pokemon:
    #data
    attackNumber = 3
    pokeTypes = ("water", "fire", "grass")
    typeMaps = {
        "fire": ["grass", "water"],
        "water": ["fire", "grass"],
        "grass": ["water", "fire"]
    }

    def __init__(self, name, type, health, defense, moves):
        self.name = name
        self.type = type
        self.health = health
        self.defense = defense
        self.moves = moves

    def getTypeMaps(self):
        return self.typeMaps

    def attack(self, other):
        print(self.name + " attacks " + other.name)
        maps = self.getTypeMaps()
        print (maps[self.type])
        if (maps[self.type][0] == other.type):
            print(f"{other.name} is weak to {self.name}")
            other.health -= self.getDamageNumber()
        elif (maps[self.type][1] == other.type) :
            print(f"{other.name} is strong to {self.name}")
            self.health -= other.getDamageNumber()
        else:
            print(other.name + " is stronger in " + self.type)
            self.health -= other.getDamageNumber()
        print(f"{self.name} health: {self.health}")
        print(f"{other.name} health: {other.health}")

    def battle(self, other, move):
        print(self.name + " battles " + other.name + " with " + move.name)
        randomMove = random.choice(other.moves)
        print(f"{other.name} uses {randomMove.name}")
        maps = self.getTypeMaps()
        if (maps[move.type][0] == randomMove.type):
            print(f"{other.name} is weak to {self.name}")
            other.health -= self.getDamageNumber(move)
        elif (maps[move.type][1] == randomMove.type) :
            print(f"{other.name} is strong to {self.name}")
            self.health -= other.getDamageNumber(move)
        else:
            print(self.name + " is stronger in " + self.type)
            self.health -= other.getDamageNumber(move)
        print(f"{self.name} health: {self.health}")
        print(f"{other.name} health: {other.health}")

    def advance(self):
        print(self.name + " advances")

    def retreat(self):
        print(self.name + " retreats")

    #damage = randomDamageNumber + move.damage
    def getDamageNumber(attacker):
        randomDamageNumber = random.randint(1, 5) + attacker.attackNumber
        print(f"{attacker.name} gives {randomDamageNumber} damage")
        return randomDamageNumber

    def getDamageNumber(self, move):
        randomDamageNumber = random.randint(1, 5) + move.damage
        print(f"{move.name} gives {randomDamageNumber} damage")
        return randomDamageNumber

class Move:
    def __init__(self, name, damage, type, super):
        self.name = name
        self.damage = damage
        self.type = type
        self.super = super
    def getDamage(self):
        return self.damage

acidMove = Move("acid", 3, "water", False)
auraSphere = Move("aura", 2, "grass", False)
aeroblast = Move("aeroblast", 2, "fire", False)
megaPunch = Move("megaPunch", 5, "fire", True)

cpuType = random.choice(Pokemon.pokeTypes)
cpuName = random.choice(["FurLizardMon", "GreyEagleMon", "JetiMon"])
cpuPoke = Pokemon(cpuName, cpuType, 10, 10, [acidMove, auraSphere, aeroblast, megaPunch ])

myName = input("Please enter your pokemon name: ")
myType = input("Please select your pokemon type: water, fire, grass ")
while (myType not in Pokemon.pokeTypes):
    myType = input("Please select your pokemon type: water, fire, grass ")
myPoke = Pokemon(myName, myType, 10, 10, [acidMove, auraSphere, aeroblast, megaPunch])
while True:
    if (cpuPoke.health <= 0):
        print(f"{cpuPoke.name} is dead, you win")
        break
    elif (myPoke.health <= 0):
        print(f"{myPoke.name} is dead, game over")
        break
    myMove = input("Please select your move: acid, aura, aeroblast, megaPunch")
    while (myMove not in [move.name for move in myPoke.moves]):
        myMove = input("Please select your move: acid, aura, grasaeroblast, megaPunch ")
    for move in myPoke.moves:
        if (move.name == myMove):
            myPoke.battle(cpuPoke, move)

```
