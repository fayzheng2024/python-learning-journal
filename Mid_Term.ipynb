```python
"""
1. make reservation. -> show adventure menu -> user input -> reservation file
2. view reservation
3. cancel reservation -> rename file to _canceled
"""

from datetime import datetime # importing datetime for use in textfile
import os # importing os
import random # importing random for random reservation number
from collections import defaultdict # importing defaultdict for use in calculate ticket totals


currentDate = datetime.now()
# get adventure name
def getAdventureName(adventure):
  if adventure == "A":
        return(f"Adventure Boof Ghost House")
  elif adventure == "B":
        return(f"Adventure Gilroy Flower Garden")
  elif adventure == "C":
        return(f"Adventure Cosmo Drink River")
  else:
        return(f"Invalid adventure")

#checkout will write the reservation info to a file and generate a reservation number
def checkout(nameStr, phoneStr, emailStr, selArray, adventures):
  resNumber = random.randint(100000, 999999)
  fileName = f"res_{resNumber}.txt"
  with open(fileName, 'w') as file:
    file.write(f"Reservation Number: {resNumber}\n")
    file.write(f"Name: {nameStr}\n")
    file.write(f"Phone: {phoneStr}\n")
    file.write(f"Email: {emailStr}\n")
    file.write("Selected Adventures:\n")
    total = 0
    #The 1st version is using array, 2nd version using dict
    """
    for adventure in selArray:
        file.write(f"{adventure}\n")
    """
    for adventure, tickets in adventures.items():
        adStr = getAdventureName(adventure)
        file.write(f"{adStr}: {tickets} tickets\n")
    file.close()
  print(f"Reservation {resNumber} saved to file: {fileName}")

def makeReservation():

  # Get date input from the user
  while True :
    dateStr = input("Enter your reservation date (YYYY-MM-DD): ")

    # Convert the string to a datetime object
    # check if it's valid datetime format
    try:
        dateObj = datetime.strptime(dateStr, "%Y-%m-%d")
        if dateObj < currentDate:
            print("Invalid date. Please enter a date in the future.")
        else:
            print("Date:", dateObj.date())
            break
    except ValueError:
        print("Invalid date format. Please use YYYY-MM-DD.")
        exit

  print("Welcome to Adventure City!")
  nameStr = input("Enter your name: ")
  phoneStr = input("Enter your phone number: ")
  emailStr = input("Enter your email: ")
  selArray  = []
  adventures = defaultdict()
  while True :
    print("Select an adventure: ")
    print(  "A.  Adventure Boof Ghost House")
    print(  "B.  Adventure Gilroy Flower Garden")
    print(  "C.  Adventure Cosmo Drink River")
    print(  "99. Check Out")
    print(  "-99. exit")
    adventureStr = input("Enter your adventure selection: ").upper()
    if(adventureStr == 'A' or adventureStr == 'B' or adventureStr == 'C'):
      try:
        tickets = int(input("Enter number of tickets: "))
      except ValueError:
        print("Invalid input. Please enter a number.")
        continue
      if tickets > 0:
        selArray.append(getAdventureName(adventureStr)+ ": " + str(tickets) + " tickets")
        if adventureStr in adventures:
          adventures[adventureStr] += tickets
        else:
          adventures[adventureStr] = tickets
        print(adventures)
        #break
      else:
        print("Invalid number of tickets. Please enter a positive number.")
    elif adventureStr == "99":
      checkout(nameStr, phoneStr, emailStr, selArray, adventures)
      break
    elif adventureStr == "-99":
      exit()
  return

# Given a reservation number, look for the reservation info
def viewReservation():
    resNumber = input("Enter reservation number: ")
    fileName = f"res_{resNumber}.txt"
    if os.path.exists(fileName):
      print(f"The file '{fileName}' exists.")
      with open(fileName, 'r') as file:
          content = file.read()
          print(content)
    else:
        print(f"The reservation '{resNumber}' does not exist.")

# Given the reservation number, look for the reservation file,
# if file exists then rename it, otherwiser error out
def cancelReservation():
    resNumber = input("Enter reservation number: ")
    fileName = f"res_{resNumber}.txt"
    canceledFileName = f"res_{resNumber}_canceled.txt"
    if os.path.isfile(fileName):
        os.rename(fileName, canceledFileName)
        print(f"The reservation '{resNumber}' has been cancelled.")
    else:
        print(f"The reservation '{resNumber}' does not exist.")
    pass

# Main program action menu
print("Welcome to Adventure City!")
print("Please select an option:")
print("1. Make a reservation")
print("2. View a reservation")
print("3. Cancel a reservation")
action = input("Enter your action: ")

if action == "1":
  makeReservation()
  pass
elif action == "2":
  viewReservation()
  pass
elif action == "3":
  cancelReservation()
  pass
else:
  print("Invalid action. Please try again.")
```
